#### Mann-Whitney U test

```python
stats.mannwhitneyu(SVM_recall, SNN_recall, method='exact', alternative='greater')
```

U stat - alpha 0.05 - greater 
- acc (statistic=31.0, pvalue=0.022475621351224686)
- prec (statistic=18.0, pvalue=0.5313852813852814)  -> SVM not higher  
- recall (statistic=36.0, pvalue=0.0010822510822510823)  
- f1 (statistic=36.0, pvalue=0.0010822510822510823)  

#### Wilcoxon signed rank test

```python
#perform the Wilcoxon-Signed Rank Test
stats.wilcoxon(SNN_acc, ANN_acc, method='exact', alternative='greater')
```

W  stat - alpha 0.05 - greater
- acc (statistic=21.0, pvalue=0.015625)
- prec  (statistic=20.0, pvalue=0.03125)
- recall (statistic=21.0, pvalue=0.015625)
- f1 (statistic=21.0, pvalue=0.015625)

#### Mean, SD and Median

```python
SNN_mean = np.mean(SNN_acc)
ANN_mean = np.mean(ANN_acc)
SVM_mean = np.mean(SVM_acc)

SNN_med = np.median(SNN_acc)
ANN_med = np.median(ANN_acc)
SVM_med = np.median(SVM_acc)

SNN_sd = np.std(SNN_acc)
ANN_sd = np.std(ANN_acc)
SVM_sd = np.std(SVM_acc)

print(SNN_mean, ANN_mean, SVM_mean)
print(SNN_med, ANN_med, SVM_med)
print(SNN_sd, ANN_sd, SVM_sd)
```

```python
SNN_prec_mean = np.mean(SNN_prec)
ANN_prec_mean = np.mean(ANN_prec)
SVM_prec_mean = np.mean(SVM_prec)

SNN_prec_med = np.median(SNN_prec)
ANN_prec_med = np.median(ANN_prec)
SVM_prec_med = np.median(SVM_prec)

SNN_prec_sd = np.std(SNN_prec)
ANN_prec_sd = np.std(ANN_prec)
SVM_prec_sd = np.std(SVM_prec)

print(SNN_prec_mean, ANN_prec_mean, SVM_prec_mean)
print(SNN_prec_med, ANN_prec_med, SVM_prec_med)
print(SNN_prec_sd, ANN_prec_sd, SVM_prec_sd)
```
